# 3D Printer ToolHead Detection

### Objective

Project Goal Our project focuses on building a computer vision model that can identify and track the toolhead of a 3D printer. We wanted to create a reliable way to monitor the printer's movements, which is the first step toward catching print failures

**The Challenges We Faced** Since this is a specific niche, we couldnâ€™t find any pre-existing datasets of 3D printer components online. To solve this, we had to build our own dataset from scratch:

**Data Sourcing:** We scraped frames from various YouTube videos to get a wide range of printer models and different lighting setups. This gave us a "real-world" variety.

**Manual Annotation:** To ensure the model learned correctly, we used AnyLabeling to manually draw bounding boxes around the toolhead in every image. While this was time-consuming, it allowed us to maintain high quality and accuracy for our "ground truth" data.

Image Preprocessing: To prepare the data for training, we resized all of our scraped images to a standard 224x224 resolution. This step was essential to keep the model efficient and compatible with standard deep learning architectures while still keeping enough detail to recognize the toolhead.
